```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[['age','affordibility']],df.bought_insurance,test_size=0.2, random_state=25)
```

```python
len(x_train)
```

```python
# scaling of data
# casue age is not in the same range as affordibility
# from sklearn.preprocessing import StandardScaler
x_train_scaled = x_train.copy()
x_train_scaled['age'] = x_train_scaled['age'] / 100

x_test_scaled = x_test.copy()
x_test_scaled['age'] = x_test_scaled['age'] / 100
```

```python
# now we can create a neural network model


modle = keras.Sequential([
    keras.layers.Dense(1,input_shape = (2,), activation= 'sigmoid', kernel_initializer = 'ones', bias_initializer='zeros')
])

# input_shape = (2,)  - this means we have 2 features in our input data
# 1 - this means we have 1 output in our model
# activation= 'sigmoid' - this means we are using sigmoid activation function
# kernel_initializer = 'ones' - this means we are initializing the weights-(w) of the model with ones
# bias_initializer='zeros' - this means we are initializing the bias-(b) of the model with zeros
```

```python
# now we can compile  the model

modle.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# optimizer='adam' - this means we are using adam optimizer
# loss='binary_crossentropy' - this means we are using binary crossentropy loss function its more efficient than that of other loss functions for binary classification problems as it suits best for gradeeint descent
# metrics=['accuracy'] - this means we are using accuracy as a metric to evaluate the model
```

```python
modle.fit(x_train_scaled,y_train,epochs=5000,batch_size=32, validation_data=(x_test_scaled,y_test),)
#  batch_size=32, validation_data=(x_test_scaled,y_test)
```

```python
# now we can evaluate the model
loss, accuracy = modle.evaluate(x_test_scaled, y_test)
print(f'Loss: {loss}, Accuracy: {accuracy}')
```

# oftenly asked in interviews

```python
def sigmoid_numpy(x):
    return 1/(1+np.exp(-x))

sigmoid_numpy(np.array([0.47, 0.18,0]))
```

```python
# def log_loss(y_true, y_predicted):
#     epsilon = 1e-15
#     y_predicted_new = [max(i,epsilon) for i in y_predicted]
#     y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
#     y_predicted_new = np.array(y_predicted_new)
#     return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def log_loss(y_true, y_pred):
    epsilon = 1e-15
    y_pred_n = [max(i,epsilon) for i in y_pred]
    y_pred_n = [min(i,epsilon) for i in y_pred_n]

    y_pred_n = np.array(y_pred_n)
    return -np.mean(y_true *np.log(y_pred_n)+(1-y_true)*np.log(1-y_pred_n))
```

```python
def gradeint_descent(age, affordiblity,y_true, epochs,loss_thresold):
    w1 = w2 = 1
    b = 0
    rate = 0.5
    n = len(age)

    for i in range(epochs):
        weighted_sum = w1*age + w2*affordiblity + b
        y_pred = sigmoid_numpy(weighted_sum)

        loss = log_loss(y_true,y_pred)

        w1d = (1/n)*np.dot(np.transpose(age),(y_pred-y_true)) 
        w2d = (1/n)*np.dot(np.transpose(affordiblity),(y_pred-y_true)) 

        bias_d = np.mean(y_pred-y_true)
        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        b = b - rate * bias_d

        print (f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{b}, loss:{loss}')

        if loss<=loss_thresold:
            break

    return w1, w2, b
```

```python
gradeint_descent(x_train_scaled['age'],x_train_scaled['affordibility'],y_train,1000, 0.4631)
```
