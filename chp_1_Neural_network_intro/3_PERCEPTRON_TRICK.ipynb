# 2. Percoeptron function

this will return us weights, bias

```python
def perceptron(x,y):
    # the bellow line will add 1 to the curent coordiates 1,x,y
    x = np.insert(x,0,1,axis=1)

    # array of weights, ahving shape as column of x i.e 3
    weights = np.ones(x.shape[1])

    # learnin rate
    lr = 0.1


    # epochs -> 1000
    for i in range(1000):

        #  selectign a ranodo int form 1,100
        j = np.random.randint(0,100)


        y_hat= step(np.dot(x[j],weights))

        weights = weights+lr*(y[j]-y_hat)*x[j]

    return weights[0], weights[1:]
    # intercet, coefficient
```
